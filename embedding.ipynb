```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Sentenses for example
sent=['the glass of milk',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good',]
```

```python
## define the vocabulary size
voc_size=10000
```

```python
## One hot representation
one_hot_rep=[one_hot(words,voc_size) for words in sent]
one_hot_rep
```

```python
## word embedding reprsentation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
embedding_docs=pad_sequences(one_hot_rep,padding="pre",maxlen=sent_length)
print(embedding_docs)
```

```python
## feauture representation
dim=10
```

```python
##model creation
model=Sequential()
model.add(Embedding(voc_size,dim,input_shape=(sent_length,)))
model.compile(optimizer='adam',loss='mse')
```

```python
print(model.summary())
```

```python
model.predict(embedding_docs)
```
